```python
#1.	Write a Python program to check if the given number is a Disarium Number?
```

disarium number.png

```python
number = int(input("enter a No: "))

original_number = number 

power = len(str(number))  # max power 
sum = 0

while number > 0 :
  rem = number % 10  # individual digit
  sum = sum + (rem**power)
  number = number // 10
  power = power - 1

if sum == original_number :
  print(f"{original_number} is disarium number")
```

```python
def disnum():
  number = int(input("enter a No: "))

  original_number = number 

  power = len(str(number))  # max power 
  sum = 0

  while number > 0 :
    rem = number % 10  # individual digit
    sum = sum + (rem**power)
    number = number // 10 # decreasing the Number for next iteration
    power = power - 1

  if sum == original_number :
    print(f"{original_number} is disarium number")
  else:
    print(f"{original_number} is not a disarium number")
```

```python
n2 = int(input("enter a number: "))
x = n2   # x is a temporary number
while x >=10:
  sum = 0
  while x > 0:
    rem = x % 10  #individual digit
    #print("remainder is: ",rem)
    sum = sum + rem**2  #summation
    x = x // 10
    #print("qoetinent is: ",x)      #reducing the no of digit in number
  print("sum: ",sum)
  x = sum    #checking for single digit

if x == 1 :
  print(n2,"is a happy number")
else :
  print(n2,"is a not happy number")

```

```python
def harshd_number():

  n5 = int(input("enter a number: "))

  orig_number = n5

  sum = 0
  while n5>0 :
    rem = n5 % 10
    sum = sum + rem
    n5 = n5 // 10

  if orig_number % sum == 0 :
    print(f"{orig_number} is a Harshad Number")

  else:
    print(f"{orig_number} is a not a Harshad Number")
    


```
